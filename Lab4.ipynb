```python
# Step 1: Calculate power
step1 = 3 ** 2
print("Step 1: 3 ** 2 =", step1)
# Step 2: Multiply
step2 = 2 * step1
print("Step 2: 2 * 3 ** 2 =", step2)
# Step 3: Divide
step3 = 8 / 4
print("Step 3: 8 / 4 =", step3)
# Step 4: Final calculation
x = 10 + step2 - step3
print("Step 4: 10 + 2 * 3 ** 2 - 8 / 4 =", x)
```

```python
n = int(input("enter a number"))
rev = 0
# Loop to reverse digits
while n > 0:
    rev = rev*10 + n%10
    n //= 10
print(rev)
```
