```python
# Taking decimal input
n = int(input("Enter number: "))
# Conversions
print("Decimal Value:",n)
print("Binary:", bin(n))
print("Octal:", oct(n))
print("Hexadecimal:", hex(n))
```

```python
# Taking float as string
s = input("Enter float value: ")
# Converting and adding
num = float(s)
print(num + 10.5)
```
