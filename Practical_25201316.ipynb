### 2.Write a program that prompts user to enter a number. If the number is equal to 99, print“Congratulations”. If the number is less than 99, print—enter again and aim higher else print enter again a lower number. The program should run until the user guesses the correct the number that is 99.

```python
#Q2
while True:
    num=int(input("Enter number: "))
    if(num==99):
        print(f"{num}: Congratulations") 
        break
    elif(num<99):
        print(f"{num}: Enter again and aim higher")
    else:
        print(f"{num}: Enter again a lower number")
        
```

### 3.Encode a string based on first occurrence pattern. Input:paper, Output: 0.1.0.2.3

```python
#Q3
s = input("Enter a string to encode: ")
print(s)
traversed = [] # to check the characters traversed
encoded= [] # to encode the position no assigned to the character
for ch in s:
    f=False
    pos=0
    for i in range(len(traversed)):
        if traversed[i]==ch:
            f=True
            pos=i
            break
    if not f:
        traversed.append(ch)
        pos=len(traversed)-1
    encoded.append(str(pos))# append the position which was assigned to the letter to the ecnoded list
print(".".join(encoded))

```
